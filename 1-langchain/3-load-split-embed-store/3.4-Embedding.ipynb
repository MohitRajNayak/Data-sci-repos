### embedding on speech.txt openAI

```python
# Load
from langchain_community.document_loaders import TextLoader
doc = TextLoader('speech.txt').load()   # we are providing the path and initializing the loader
doc
```

```python
# Split
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_text = text_splitter.split_documents(doc)
final_text
```

```python
## Embeded
from langchain_openai import OpenAIEmbeddings
embedding_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

## vector store

```python
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(documents=final_text,embedding=embedding_1024)
```

using free tier so billing error 

```python
query = "Langchain for gen ai"
retrive = db.similarity_search(query)
print(retrive)
```
