# FAISS

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

docs = TextLoader('speech.txt').load()
text_split = CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docs_split = text_split.split_documents(docs)
```

```python
docs_split
```

```python
embedding = OllamaEmbeddings(model="mxbai-embed-large")
db =FAISS.from_documents(docs_split,embedding)
db
```

```python
# query

query = "what does the speaker believe is the main reason the united States should enter the war?"

docs = db.similarity_search(query)
docs
```

```python
docs[0]
```

Retriver

```python
retriver = db.as_retriever()
docs = retriver.invoke(query)
docs[0].page_content
```

Similarity search with score

```python
docs_and_score = db.similarity_search_with_score(query)
```

```python
docs_and_score
```

Similarity search by vector

```python
embedding_vector = embedding.embed_query(query)
embedding_vector
```

```python
docs = db.similarity_search_by_vector(embedding_vector)
docs
```

Store VectorDB in local and leading

```python
# Save in local
db.save_local("faiss_index")
```

```python
# Load from local same file faiss
new_db = FAISS.load_local("faiss_index",embedding,allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query)
docs
```

# Chroma DB

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

docs = TextLoader('speech.txt').load()
text_split = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
embedding = OllamaEmbeddings(model='mxbai-embed-large')
```

```python
docs_split = text_split.split_documents(docs)
docs_split
```

```python
db = Chroma.from_documents(documents=docs_split,embedding=embedding)
db
```

```python
# query

query = "what does the speaker believe is the main reason the united States should enter the war?"

docs_query = db.similarity_search(query)
docs_query[0].page_content
```

```python
# save local
db = Chroma.from_documents(docs_split,embedding,persist_directory="./chroma_db")
```

```python
# Load db

db2 = Chroma(persist_directory="./chroma_db",embedding_function=embedding)
docs = db2.similarity_search_with_score(query)
docs[0]
```

retriver option

```python
retriver = db.as_retriever()
retriver.invoke(query)[0]
```
